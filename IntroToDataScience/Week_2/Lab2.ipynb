Last class, we covered hypotheses testing as well as t-test to compare two population means. In this lab, you will run a t-test yourself, report the results, and write whether the null hypothesis is rejected or not.  

Here is the summary of what we learned about hypotheses testing so far:

**Hypothesis testing** starts with an assumption that the null hypothesis (H0) is true and it attempts to find evidence to reject it. alternative hypothesis (Ha) represents the opposite of the H0 and it is the hypothesis we are trying to prove.

**The significance level**, denoted by alpha, represents the probability of rejecting the H0, assuming it is true (and it is usually 0.05).

**The p-value** - the probability of obtaining test results, assuming the H0 is true. A small p-value indicates that the probability of obtaining the observed results is very low, given that the H0 is correct. in other words, it means it would be very unlikely that we would get the results supporting Ha when H0 is actually true. 

**A t-test** is used to determine if there is a significant difference between the means of the two groups.

If the obtained p-value is lower than the significance level (α), we can conclude that **there is a statistically significant difference between the two means.**
