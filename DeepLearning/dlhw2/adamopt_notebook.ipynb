12. (coding) Adam optimization

```python
import numpy as np
```

First, I began by defining the object function, which is a quadratic function, while also setting up my plt plots

```python
def f(w):
    return (w - 3) ** 2

```

Next, I add the derivative calculation and define my gradient function. When w = 3 , we know thr gradient is at the minimum


```python
def grad_f(w):
    """Gradient of f(w) = (w - 3)^2"""
    return 2 * (w - 3)


test_points = [0, 3, 10]
print("Gradient at different points:")
for w in test_points:
    print(f"  w={w:2.0f}: grad={grad_f(w):6.2f}")
```

Then I set up the Adam parameters (setting the w to 10)

```python

w = 10.0                  


lr = 0.1                  
beta1 = 0.9               
beta2 = 0.999            
eps = 1e-8               


m = 0.0                   
v = 0.0              
t = 0          

print(f"Starting optimization from w={w}")
print(f"Hyperparameters: lr={lr}, β1={beta1}, β2={beta2}")
```

I add the moment updates (m and v) to the loop (first moment exponentially weighted avg of gradients, the second avg of squared gradients (like RMSProp)), as well as edit the bias correction to see what has been

Finally we plug them into the param updates and see what has been added


```python
# Track optimization history
history = {'step': [], 'w': [], 'f_w': [], 'grad': [], 'm': [], 'v': []}

# Reset for demonstration
w = 10.0
m = 0.0
v = 0.0
t = 0

print("\nOptimization Progress:")
print("-" * 70)

for step in range(1, 51):
    t += 1
    
    # Compute gradient at current point
    g = grad_f(w)
    

    m = beta1 * m + (1 - beta1) * g
    v = beta2 * v + (1 - beta2) * (g ** 2)
    

    m_hat = m / (1 - beta1 ** t)
    v_hat = v / (1 - beta2 ** t)
    
    w = w - lr * m_hat / (np.sqrt(v_hat) + eps)
    
    # Track history
    history['step'].append(step)
    history['w'].append(w)
    history['f_w'].append(f(w))
    history['grad'].append(g)
    history['m'].append(m)
    history['v'].append(v)
    
    # Print selected steps
    if step in [1, 2, 3, 5, 10, 20, 50]:
        print(f"step {step:2d} | w={w:.6f} | f(w)={f(w):.6f} | grad={g:.6f}")

print(f"\nFinal result: w={w:.6f} (optimal: 3.0)")
print(f"Final loss: f(w)={f(w):.8f} (optimal: 0.0)")
```
